```python
import pandas as pd

path = '/home/paulo/git/reconhecimento-padroes/projeto-final/data/pedalboard-plugin-bag-of-words.csv'
presets = pd.read_csv(path, index_col=['index', 'id'])

# Remove guitar_patches.id column
presets = presets.reset_index(level='id', drop=True)
# Remove never unused plugin
# > presets.T[pedalboards_filtered.sum() == 0]
# 9
presets['9'] = 0
presets['107'] = 0
presets = ((presets > 0) * 1)
presets = presets[presets.T.sum() > 3]

print(len(presets))
presets.head()
```

```python
def conditional_probability(dataset, variables):
    groupby = dataset.groupby(variables).sum()

    return groupby / groupby.sum()

conditional = conditional_probability(presets, ['1', '3'])
valores = (1, 1)
conditional.loc[valores].sort_values(ascending=False)
```
